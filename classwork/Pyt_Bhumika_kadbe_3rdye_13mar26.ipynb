```python
#Create a dictionary storing product purchase frequency.
purchases = ["Laptop","Phone","Laptop","Tablet","Phone","Laptop"]
print(purchases)
print(len(purchases))

frequency = {}
for purchase in purchases:
    if purchase in frequency:
        frequency[purchase] += 1
    else:
        frequency[purchase] = 1
print(frequency)

# printing the dictionary
for product, count in frequency.items():
    print(f"{product}: {count}")
    
# identify the most purchased product
most_purchased_product = max(frequency, key=frequency.get)
print("Most Purchased Product: ", most_purchased_product)
```

```python
day1 = ["U101","U102","U103","U104"]
day2 = ["U103","U104","U105","U106"]

#identify common visitor
common_visitor =[]
for i in day1:
    if i in day2:
        common_visitor.append(i)
print("Common Visitors: ", common_visitor)

unique_visitors =[]
for i in day1:
    if i in day2:
        continue
    else:
        unique_visitors.append(i)   

for i in day2:
    if i in day1:
        continue
    else:
        unique_visitors.append(i)

print("Unique Visitors: ", unique_visitors)

only_day1_visitors = []
for i in day1:
    if i in day2:
        continue
    else:
        only_day1_visitors.append(i)  
print("Only in day1 : ", only_day1_visitors)
    
        
```

```python
employees = [
("E101","Data Analyst"),
("E102","Data Engineer"),
("E103","Data Analyst"),
("E104","ML Engineer"),
("E105","Data Engineer")
]
# Create a dictionary to count the number of employees in each department
department_count = {}
for emp_id, department in employees:
    if department in department_count:
        department_count[department] += 1
    else:
        department_count[department] = 1
print("Department Count: ", department_count)

#for no ofemploye per department
for department, count in department_count.items():
    print(f"{department}: {count}")
    
#identify trhe department with highest no of employees
most_employees_department = max(department_count, key=department_count.get)
print("Department with the highest number of employees: ", most_employees_department)
```

```python
team_skills = {
"Alice": ["Python","SQL","Tableau"],
"Bob": ["Python","Spark","SQL"],
"Charlie": ["Python","Machine Learning","SQL"]
}

#unique set of skills
unique_skills = set()
for skills in team_skills.values():
    unique_skills.update(skills)
print("Unique Skills: ", unique_skills)
print(len(unique_skills))

#print the skill list in alphabetical order
sorted_skills = sorted(unique_skills)
print("Sorted Skills: ", sorted_skills)
```

```python
sales = [
("Laptop",1200),
("Phone",800),
("Laptop",1200),
("Tablet",600),
("Phone",800)
]

#  total revenue per product
revenue = {}
for product, price in sales:
    if product in revenue:
        revenue[product] += price
    else:
        revenue[product] = price

print("Total Revenue by Product: ", revenue)

# create a new dict product sorting according to product
sorted_pr = sorted(sales, key=lambda x: x[1]) 
print("Products sorted by name: ", sorted_pr)

#Identify the highest revenue generating product.
highest_revenue_product = max(revenue, key=revenue.get)
print("Highest Revenue Generating Product: ", highest_revenue_product)
```

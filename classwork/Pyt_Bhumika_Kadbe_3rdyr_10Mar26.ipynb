```python
# perform append() on list --->inseert at end without index no
list_sub = ["English", "Maths", "Science", "Marathi", "Hindi", "Sanskrit", "Maths"]
print(list_sub)
list_sub.append("Computer")
print(list_sub)

#perform insert() on list---->insert with index no
list_sub.insert(2, "History")
print(list_sub)

#checking len of list
print(len(list_sub))

#perform pop() on list--->remove last element from list (cannot use the values only index no)
list_sub.pop(2)
print(list_sub)

#count
print(list_sub.count("Maths"))

#perform remove() on list ---> remove elements with values
list_sub.remove("Maths")
print(list_sub)

#perform clear() on list---> remove all elements from list
list_sub.clear()
print(list_sub)

#making new list 
ls=[12.3, 4.5, 6.7, 8.9, 2.0, 3.4, 5.6, 1.0, 3.56, 7.8]
print(ls)
ls.sort()
print(ls) 
```

```python
# performing Different Functions in Tuple

# making new tuple
tup = ("Apple", "Banana", "Grapes", "Orange", "Pineapple", "Cherry", "Strawberry")
print(tup)

# checking length of tuple
print(len(tup))

# accessing tuple using index number
print(tup[2])

# finding index of element in tuple
print(tup.index("Grapes"))

# performing count() on tuple
print(tup.count("Grapes"))

# converting tuple into list
tup_list = list(tup)
print(tup_list)
```

```python
# Performing Different function on sets

st = {"Nagpur", "Amaravati", "Mumbai", "Pune", "Nashik", "Delhi"}
print(st)

# adding element in set --> add anywhere because sets are unordered
st.add("Bangalore")
print(st)

# adding multiple elements
st.update(["Hyderabad", "Chandrapur"])
print(st)

# checking length
print(len(st))

# pop an element
st.pop()
print(st)

# removing element
st.remove("Nashik")
print(st)

# clearing set
st.clear()
print(st)
```

```python
# creating dictionary
dict1 = {
    "name": "Bhumika",
    "age": 19,
    "course": "CSE",
    "city": "Nagpur",
    "lazy": "True"
}

print(dict1)

# add new key
dict1["college"] = "JIT"
print(dict1)

# update value
dict1["age"] = 20
print(dict1)

# access value
print(dict1["name"])

# length
print(len(dict1))

# remove key
dict1.pop("course")
print(dict1)

# get all keys
print(dict1.keys())

# get all values
print(dict1.values())

# clear dictionary
dict1.clear()
print(dict1)
```

```python
#creating list name scores
scores = [78, 40, 92, 34, 67, 88, 53, 41, 76, 29, 64, 81, 39, 55, 72, 90, 47, 68, 33, 84, 0, -5, 100, 40, 75]
print(scores)

l1=scores[0:10]
print("first 10 scores", l1)

l2=scores[-5:]
print("last five scores", l2)

l3=scores[1::2]
print("scores after 2 interval: ", l3)

pass_students =[]
failed_students =[]
for i in scores:
    if(i>=40):
        pass_students.append(i)
    else:
        failed_students.append(i)

print("Passed Students before adding 5 marks::", pass_students)
print("Failed Students before adding 5 marks: ", failed_students)

#making new csire list increase by 5
increase = []
for i in scores:
    i = i+5
    increase.append(i)

print("Increased Marks list: ",increase)

only_top =[]
for i in increase:
    if(i>=75):
        only_top.append(i)
print("Top Students More than 75 marks:",only_top)

```

```python
orders = [
("ORD001", "C101", "Laptop", 1200),
("ORD002", "C102", "Phone", 800),
("ORD003", "C101", "Laptop", 1200),
("ORD004", "C103", "Tablet", 600),
("ORD005", "C104", "Phone", 800),
("ORD006", "C102", "Headphones", 150),
("ORD007", "C105", "Laptop", 1200),
("ORD003", "C101", "Laptop", 1200),   # duplicate order
("ORD008", "C106", "Tablet", 600),
("ORD009", "C101", "Phone", 800)
]

#for printing the order list
for i in orders:
    print(i)
print("Total no of orders: ",len(orders))
```

```python
#remove duplicate records
unique_records = set(orders) 
for record in unique_records:
    print(record)
    
#printing the number of unique records
print("Total Unique Records: ",len(unique_records))
```

```python
#idedentify unique customers
unique_customers = set()
for order in orders:
    unique_customers.add(order[1])
print("Unique Customers: ",unique_customers)
```

```python
#extract product name using list comprehension
product_names = [order[2] for order in orders]
print("Product are : ", product_names)
print(len(product_names))

unique_product_name = set(product_names)
print("Unique Product Names: ", unique_product_name)
print(len(unique_product_name))
```

```python
#count how many different products are sold
total_product_sales = {}
for order in orders:
    if order[2] in total_product_sales:
        total_product_sales[order[2]] += 1
    else:
        total_product_sales[order[2]] = 1

print("Total Sales for each product: ", total_product_sales)
```

```python
#order[2] is the product name and order[3] is revenue
#print unique product with revenue
total_product_revenue = {}
for order in orders:
    product_name = order[2]
    revenue = order[3]
    if product_name in total_product_revenue:
        total_product_revenue[product_name] += revenue
    else:
        total_product_revenue[product_name] = revenue
print("Total Revenue for each product: ", total_product_revenue)
    
```

```python
#customers with no of orders
customer_order_count = {}
for order in orders:
    customer_id = order[1]
    if customer_id in customer_order_count:
        customer_order_count[customer_id] += 1
    else:
        customer_order_count[customer_id] = 1

print("Total Orders for each customer: ", customer_order_count)

#customer with maiximum orders
max_orders = max(customer_order_count.values())
top_customer = []
for customer, count in customer_order_count.items():
    if count == max_orders:
        top_customer.append(customer)
print("Customer with maximum orders: ", top_customer)
```

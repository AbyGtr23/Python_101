```python
data = input("Enter the number : ").split()

# counters
even_count = 0
odd_count = 0
invalid_count = 0

for i in data:
    try:
        num = int(i)

        if num % 2 == 0:
            print("Even number")
            even_count += 1
        else:
            print("Odd Number")
            odd_count +=1

    except:
        print(i,"is invalid, skipped")
        invalid_count += 1

print("\nFinal Report")
print("Event Numbers:",even_count)
print("odd Number:",odd_count)
print("Invalid input",invalid_count)
        
```

```python
i= input("Enter sequence of number: ")
num = i.split()
valid_count = 0
invalid_count = 0
for j in num:
    try:
        n=int(j)
        valid_count += 1
        if n%2==0:
            print(f"{n} is even number")
        else:
            print(f"{n} is odd number")

    except (TypeError,ValueError):
        print("invalid input")
        invalid_count +=1           

print("\nTotal valid numbers:", valid_count)
print("Total invalid inputs:", invalid_count)
```

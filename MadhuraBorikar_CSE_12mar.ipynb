```python
orders = [
("ORD001", "C101", "Laptop", 1200),
("ORD002", "C102", "Phone", 800),
("ORD003", "C101", "Laptop", 1200),
("ORD004", "C103", "Tablet", 600),
("ORD005", "C104", "Phone", 800),
("ORD006", "C102", "Headphones", 150),
("ORD007", "C105", "Laptop", 1200),
("ORD003", "C101", "Laptop", 1200),   # duplicate order
("ORD008", "C106", "Tablet", 600),
("ORD009", "C101", "Phone", 800)
]



# sum of total revenue


# unique customers
unique_customers = list(set(order[1] for order in orders))
print(unique_customers)

# result in dictionary with product as key and total revenue as value



#
```

```python
# stage 1:

# print all orders
for order in orders:
    print(order)

# priny total number of orders received
print(f"Total orders: {len(orders)}")

# extracting product names from orders
product_name =[]
for order in orders:
    product_name.append(order[2])       
print(product_name)
```

```python
# stage 2:

# printing unique product names
unique_product = set(orders)
for i in unique_product:
    print(i)


# printing total number of unique orders
total_unique_orders = len(unique_product)
print(f"Total unique orders: {total_unique_orders}")


# unique customers
customers = set(order[1] for order in orders)
print(customers)
```

```python
# Stage 3:
# list comprehension to extract product names
product_name_comp = [order[2] for order in orders]  
print(product_name_comp)

# set of unique product names using set comprehension
unique_product_comp = {order[2] for order in orders}
print(unique_product_comp)

# how many times each product was ordered
product_count = {}
for order in orders:
    product = order[2]
    if product in product_count:
        product_count[product] += 1
    else:
        product_count[product] = 1
print(product_count)
```

```python
# stage 4:
# total revenue
total_revenue = sum(order[3] for order in orders)
print(f"Total revenue: Rs.{total_revenue}") 

# store revenue by product in a dictionary
revenue_by_product = {}
for order in orders:
    product = order[2]
    revenue = order[3]
    if product in revenue_by_product:
        revenue_by_product[product] += revenue
    else:
        revenue_by_product[product] = revenue
print(revenue_by_product)

# print revenue per product simple
for product, revenue in revenue_by_product.items():
    print(f"Revenue for {product}: Rs.{revenue}")
```

```python
# stage 5
customer_to_no_of_orders = {}
for order in orders:
    customer = order[1]
    if customer in customer_to_no_of_orders:
        customer_to_no_of_orders[customer] += 1
    else:
        customer_to_no_of_orders[customer] = 1
print(customer_to_no_of_orders)

# maximum orders by a customer with customer id
max_customer = max(customer_to_no_of_orders, key=customer_to_no_of_orders.get)
max_orders = customer_to_no_of_orders[max_customer]
print(f"Customer {max_customer} has the most orders with a total of {max_orders}")
```

```python
# total reveneue 
total_revenue = sum(order[3] for order in orders)
print(f"Total revenue: Rs.{total_revenue}")

max_product = max(revenue_by_product, key=revenue_by_product.get)
# max_revenue = revenue_by_product[max_product]
print(f"Product with highest revenue: {max_product}")

customer_highest_orders = max(customer_to_no_of_orders, key=customer_to_no_of_orders.get)
print(f"Customer with highest orders: {customer_highest_orders}")


total_unique_customers = len(set(order[1] for order in orders))
print(f"Total unique customers: {total_unique_customers}")

total_unique_products = len(set(order[2] for order in orders))
print(f"Total unique products: {total_unique_products}")
```

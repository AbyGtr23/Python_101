```python
input_str =float(input("Enter a amount: ")) 
if input_str > 5000:    
    discount = 0.15 * float(input_str)
    amt = float(input_str) - discount
    print(f"You got 15% discount! Your discount amount is:{discount}")
elif input_str > 1000:   
    discount = 0.1 * float(input_str)
    print(f"You got 10% discount! Your discount amount is:{discount}")
elif input_str > 500:   
    discount = 0.06 * float(input_str)
    print(f"You got 6% discount! Your discount amount is:{discount}")
else:
    print("You are not eligible for any discount.")

print(f"Your final amount is: {amt}")
```

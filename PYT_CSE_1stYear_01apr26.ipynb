```python
def New_demo():
    inp=input().lower()
    print(inp)
    print("OPTIONS: iNFO, GREET, EXIT")
    while inp != "exit":
        if inp == 'info' or inp == 'greet':
            print("This is a demo")
        else:
            print("Wrong input! Enter valid options")
        break
        


New_demo()


    
```

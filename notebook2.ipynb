```python
choice = ''
while choice != 'Exit':
    print("Please choose an option: Greet, Info, Exit")
    choice = input("Enter You Choice:").lower()
    if choice == 'greet':
        print("Hello! Nice to meet you.")
    elif choice == 'info':
        print("This is a simple command-line program.")
    elif choice == 'exit':
        print("Goodbye!")
    else:
        print("Invalid choice. Please try again.")
```

```python
list_A = [1,2,3,4,5,6,7,8]
search_item = 5
for i in list_A:
    if i == search_item:
        break
    else:
        print(i)
```

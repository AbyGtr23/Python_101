```python
try:
    amount = float(input("Enter the Amount: "))
except:
    print("Invaib Input")
```

```python
if amount >= 5000:
    discount = 0.15
elif amount >=1000:
    discount = 0.10
elif amount >= 600:
    discount = 0.06
else:
    print("Invalid!")
```

```python
dis_amount = amount * discount
final_amount = amount - dis_amount
```

```python
print("\n------ BILL SUMMARY ------")
print("Original Amount :", amount)
print("Discount Applied:", discount * 100, "%")
print("Discount Amount :", dis_amount)
print("Final Payable   :", final_amount)
```

```python
# Python program to process a sequence of numbers

# Ask user for input
user_input = input("Enter numbers separated by spaces: ")

# Split input into list
items = user_input.split()
print(items)

# Initialize counters
valid_count = 0
invalid_count = 0

# Iterate through each item
for item in items:
    try:#using try-except
        num = int(item)  # converting to integer
        valid_count += 1
        if num % 2 == 0:
            print(f"{num} -> Even number")
        else:
            print(f"{num} -> Odd number")
    except ValueError:
        invalid_count += 1
        print(f"{item} -> Invalid input detected")

# Final result
print("\nresult:")
print(f"Total valid numbers: {valid_count}")
print(f"Total invalid inputs: {invalid_count}") 
```

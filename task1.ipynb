```python

# Madhura 12 Feb 2026
# discount calculator
original_price = float(input("Enter the original price of item: "))


try :
    original_price = float(original_price)
except ValueError:
    print("Please enter a valid number for the original price.")
    exit()

total_discount = 0
if original_price > 500 :
    total_discount = (6*original_price)/100
elif original_price >1000:
    total_discount = (10*original_price)/100
elif original_price >5000:
    total_discount = (15*original_price)/100 

print(f"total discount: {total_discount}")

final_price = original_price-total_discount

print(f"total bill: {final_price}")
                         

```

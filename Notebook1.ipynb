```python
input_str =float(input("Enter a amount: ")) 
if input_str > 1200:    
    discount = 0.2 * float(input_str)
    amt = float(input_str) - discount
    print(f"You got 20% discount! Your discount amount is:{discount}")
    print(f"Your final amount is: {amt}")
elif input_str > 800:   
    discount = 0.1 * float(input_str)
    print(f"You got 10% discount! Your discount amount is:{discount}")
    print(f"Your final amount is: {amt}")
else:
    print("You are not eligible for any discount.")
```

```python
# create a CLI based  discount calculator 
print("Welcome to the discount calculator!")
amount = float(input("Enter the original amount: "))
if amount > 500:
     discount = 0.15

elif amount > 1000:
     discount = 0.10

elif amount > 5000:
      discount = 0.06

else:
     print("Invalid amount")
 

print(f"Total discount : {discount*amount}\nTotal amount after discount:{amount-discount*amount}")
```

```python
tup_1 = ('a', 'b', 'c')
#Add an element 'd' to the above tuple.

t_lst = list(tup_1)
print(t_lst)

#Converting the tuple to list and then adding the element

t_lst.append('d')

#Convert it back to tuple

t_lst = tuple(t_lst)
print(t_lst)
```

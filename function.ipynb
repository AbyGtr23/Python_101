define list and define variable name search item 

```python
list1=[1,2,3,4,5,6]
searchitem=input("Enter the item to search: ")
for i in list1:
    if i== searchitem:
        break
    else:
        print(i)
```

```python
list1=[1,2,3,4,5,6]
searchitem=input("Enter the item to search: ")
for i in list1:
    if i== searchitem:
        continue
    else:
        print(i)
```

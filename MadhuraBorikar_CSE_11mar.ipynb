```python
scores = [78, 40, 92, 34, 67, 88, 53, 41, 76, 29, 64, 81, 39, 55, 72, 90, 47, 68, 33, 84, 0, -5, 100, 40, 75]

score_passed = []
score_failed =[]


 
print(scores[0:10])

print(scores[-5:])

print(scores[1::2])


# checking if passed or failed
for score in scores:
    if score >= 40:
        score_passed.append(score)
    else:
        score_failed.append(score)


# Incresed by 5
increased_by_5 = []
for score in scores:
    score+=5
    increased_by_5.append(score)


greater_than_75 =[]
# Greater than 75
for score in scores:
    if score > 75:
        greater_than_75.append(score)




 
print(f"passed student : {score_passed}")
print(f"failed student : {score_failed}")
print(f"Increased by 5 marks: {increased_by_5}") 
print(f"Top Student greater than 75: {greater_than_75}")
```

```python
amt = float(input("Enter the amt : "))

if amt >= 5000:
    dis = 0.15

elif amt >=1000:
    dis = 0.10

elif amt >=500:
    dis = 0.05

else:
    print("Invalid amount")

discount = amt * dis
final_amt = amt - discount

print("Total Discount:",discount)
print("Total amount after discount : " ,final_amt)
```

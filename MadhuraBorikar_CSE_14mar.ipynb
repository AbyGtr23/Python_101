```python
day_visits = ["U101","U102","U103","U101","U104","U102","U105","U103"]
# unique visitors
unique_visits = set(day_visits)
print(unique_visits)   

# total unique visitors
total_unique_visitors = len(unique_visits)
print(f"total_unique_visitors : {total_unique_visitors}") 

# list of unique visitors
unique_visits_list = list(unique_visits)
print(unique_visits_list)
```

```python
# Data:
day1 = ["U101","U102","U103","U104"]
day2 = ["U103","U104","U105","U106"]


# common visitors
for i in day1:
    if i in day2:
        print(f"{i} is a common visitor")

# only present in day1
for i in day1:
    if i not in day2:
        print(f"{i} is a unique visitor in day1")

# unique visitors in both days

unique_visitors_both_days = set(day1) | set(day2)
print(f"Unique visitors in both days: {unique_visitors_both_days}")
```

```python
# Scenario:
# An e-commerce company wants to count how many times each product was purchased.
# Data:
purchases = ["Laptop","Phone","Laptop","Tablet","Phone","Laptop"]

# Create a dictionary storing product purchase frequency.
product_count = {}
for product in purchases:
    if product in product_count:
        product_count[product] += 1
    else:
        product_count[product] = 1
print(product_count)


# Print the dictionary.
for product, count in product_count.items():
    print(f"{product}: {count} purchases")

# Find the product with the most purchases and print it along with the number of purchases.   
max_product = max(product_count, key=product_count.get)
max_purchases = product_count[max_product]
print(f"Product {max_product} has the most purchases with a total of {max_purchases}")  
```

```python
employees = [
("E101","Data Analyst"),
("E102","Data Engineer"),
("E103","Data Analyst"),
("E104","ML Engineer"),
("E105","Data Engineer")
]


# department-wise employee count
employees_dict ={}
for emp_id, role in employees:
    if role in employees_dict:
        employees_dict[role]+=1
    else:
        employees_dict[role] = 1
print(employees_dict)

# employee count for each department
for department, count in employees_dict.items():
    print(f"{department}: {count} employees")

# department name with the highest number of employees 

max_department = max(employees_dict, key=employees_dict.get)
max_employees = employees_dict[max_department]
print(f"Department {max_department} has the most employees with a total of {max_employees}")
```

```python
# Data:
team_skills = {
"Alice": ["Python","SQL","Tableau"],
"Bob": ["Python","Spark","SQL"],
"Charlie": ["Python","Machine Learning","SQL"]
}

unique_skills = set()
for skills in team_skills.values():
    unique_skills.update(skills)  

print(f"Unique skills in the team: {unique_skills}")

print(f"Total unique skills in the team: {len(unique_skills)}")



unique_skills_list = sorted(list(unique_skills))
print(f"List of unique skills in the team: {unique_skills_list}")
```

```python
sales = [
("Laptop",1200),
("Phone",800),
("Laptop",1200),
("Tablet",600),
("Phone",800)
]

product_revenue ={}
for product, price in sales:
    if product in product_revenue:
        product_revenue[product] += price
    else:
        product_revenue[product] = price
print(product_revenue)

for product, revenue in product_revenue.items():
    print(f"{product}: ${revenue} total revenue")


max_revenue_product = max(product_revenue, key=product_revenue.get)
max_revenue = product_revenue[max_revenue_product]
print(f"Product {max_revenue_product} generated the most revenue with a total of ${max_revenue}")
```

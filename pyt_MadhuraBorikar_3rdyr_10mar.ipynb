```python
string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")


try:
    divide = int(string1) / int(string2)
    print(f"The division of two integer is {divide}")  
    
except ZeroDivisionError:
    print("Error : Division by zero is not allowed.")

except ValueError:
    print("Error : Invalid input. Please enter valid integer.")
```

```python
numbers = input("Enter numbers (seperated by space): ")
numbers_list = numbers.split()
print(f"The list of numbers is: {numbers_list}")

valid_numbers = []
valid_count = 0
invalid_numbers = []
invalid_count = 0


for num in numbers_list:
    try:
        num_int = int(num)
        valid_numbers.append(num_int)
        valid_count += 1

        if num_int % 2 == 0:
            print(f"{num_int} is an even number.")
        else:
            print(f"{num_int} is an odd number.")
    except ValueError:
        invalid_numbers.append(num)
        invalid_count += 1

print(f"Valid numbers: {valid_numbers}")
print(f"Number of valid numbers: {valid_count}")
print(f"Invalid numbers: {invalid_numbers}")
print(f"Number of invalid numbers: {invalid_count}")
```

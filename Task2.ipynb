```python
# make a list and define a variable called search_item, print other things until search_item is found
```

```python
List1 = [1,2,3,4,5,6]
search_item = int(input("Enter : "))

for items in List1:
    if items == search_item:
        break
    else:
        print(items)
```

```python
#skip an item and print other items
```

```python
List1 = [1,2,3,4,5,6]
search_item = 5

for items in List1:
    if items == search_item:
        continue
    else:
        print(items)
```

```python
#create a list and tuple and perform operations on then
```

```python
list2 = [1,2,3,4,5]

#append
list2.append(6)
print(list2)

#insert
list2.insert(1,15)
print(list2)

#remove
list2.remove(5)
print(list2)

#pop
list2.pop(3)
print(list2)

#sort
list2.sort()
print(list2)

#reverse
list2.reverse()
print(list2)
```

```python
lst2 = [1,2,3]
lst3 = [4,5,6]

#extend
lst2.extend(lst3)
print(lst2)
```

```python
tuple1 = (10,20,30,40)

#length
print(len(tuple1))

#occurence
print(tuple1.count(20))

#indexing
print(tuple1.index(30))
```
